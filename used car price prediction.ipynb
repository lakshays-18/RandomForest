```python
df.drop(columns=['car_name','brand'],inplace=True)
```

```python
num_features = [feature for feature in df.columns if df[feature].dtype!='O']
cat_features = [feature for feature in df.columns if df[feature].dtype=='O']
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['selling_price'],axis=1)
y = df['selling_price'] 
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler
```

```python
le = LabelEncoder()
X['model']= le.fit_transform(X['model'])
```

```python
from sklearn.compose import ColumnTransformer

scaler = StandardScaler()
one_hot = OneHotEncoder(drop='first')

preprocessor = ColumnTransformer(
    [
        ('Scaler',scaler,num_features),
        ('onehot',one_hot,cat_features)
    ],remainder = 'passthrough'
)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
X_train = preprocessor.fit_transform(X_train)
```

```python
X_test = preprocessor.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
def results(actual,predicted):
    rmse = np.sqrt(mean_squared_error(actual,predicted))
    mae = mean_absolute_error(actual,predicted)
    r2 = r2_score(actual,predicted)
    return rmse,mae,r2
```

```python
models = {
    'KNN' : KNeighborsRegressor(),
    'Decision Tree' : DecisionTreeRegressor(),
    'Random Forest' : RandomForestRegressor()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)
    
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    train_rsme , train_mae , train_r2 = results(y_train,y_train_pred)
    rsme , mae , r2 = results(y_test,y_pred)
    
    print(list(models.keys())[i]," training results")
    print(f"Root Mean Squared Error: {train_rsme}")
    print(f"Mean Absolute Error: {train_mae}")
    print(f"R2 Score: {train_r2}")
    print("__________________________________________\n__________________________________________")
    print(list(models.keys())[i]," testing  results")
    print(f"Root Mean Squared Error: {rsme}")
    print(f"Mean Absolute Error: {mae}")
    print(f"R2 Score: {r2}")
    
    
```

```python
knn_params = {"n_neighbors":[2,3,5,10,15,20,50]}
rf_params = {"max_depth":[5,8,10,15,None],
             "max_features":["sqrt", "log2",None],
             "min_samples_split":[2,8,15,20],
             "n_estimators":[100,200,300,400,500]
            }

random_cv_models = [
    ('KNN',KNeighborsRegressor(),knn_params),
    ('RF',RandomForestRegressor(),rf_params)
]
```

```python
from sklearn.model_selection import RandomizedSearchCV

model_params = {}
for name,model,params in random_cv_models:
    ran = RandomizedSearchCV(estimator=model,
                             param_distributions=params,
                             n_iter=100,
                             cv=3,
                             verbose=2
                            )
    ran.fit(X_train,y_train)
    model_params[name] = ran.best_params_
    
for model in model_params.keys():
    print(f"Best Parameters for {model} are:")
    print(model_params[model])

    
    
    
```
