## Feature Engineering

```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].replace('Fe Male', 'Female')
```

```python
for i in df.columns:
    if df[i].isnull().sum()>=1:
        print(i,df[i].isnull().sum())
```

```python
df.describe()
```

```python
df['TypeofContact'].mode()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Age'].isnull().sum()
```

```python
df['TypeofContact'] = df['TypeofContact'].fillna(df['TypeofContact'].mode()[0])
```

```python
df['DurationOfPitch'] = df['DurationOfPitch'].fillna(df['DurationOfPitch'].median())
```

```python
df['NumberOfFollowups'] = df['NumberOfFollowups'].fillna(df['NumberOfFollowups'].mode()[0])
```

```python
df['PreferredPropertyStar'] = df['PreferredPropertyStar'].fillna(df['PreferredPropertyStar'].mode()[0])
```

```python
df['NumberOfTrips'] = df['NumberOfTrips'].fillna(df['NumberOfTrips'].median())
```

```python
df['NumberOfChildrenVisiting'] = df['NumberOfChildrenVisiting'].fillna(df['NumberOfChildrenVisiting'].mode()[0])
```

```python
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())
```

```python
df['MonthlyIncome'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['TotalVistin'] = df['NumberOfPersonVisiting'] + df['NumberOfChildrenVisiting']
```

```python
df
```

```python
df.drop(columns=['NumberOfPersonVisiting','NumberOfChildrenVisiting'],inplace = True)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['ProdTaken'],axis=1)
y = df['ProdTaken']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=10)
X_test
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

cat_features = X.select_dtypes(include='object').columns
num_features = X.select_dtypes(exclude='object').columns

scaler = StandardScaler()
onehot = OneHotEncoder()

preprocessor = ColumnTransformer(
    [
        ('OneHotEncoder',onehot,cat_features),
        ('StandarScaler',scaler,num_features)
    ]
)
```

```python
X_train = preprocessor.fit_transform(X_train)
```

```python
X_test = preprocessor.transform(X_test)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay,precision_score,recall_score,f1_score,roc_auc_score
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_train = model.predict(X_train)
```

```python
print("Accuracy for Training Data")
print(accuracy_score(y_train,y_pred_train))
print(precision_score(y_train,y_pred_train))
print(recall_score(y_train,y_pred_train))
print(f1_score(y_train,y_pred_train))
print(roc_auc_score(y_train,y_pred_train))
print("____________________________________________________________________________")
print("Accuracy for Test Data")
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
print(roc_auc_score(y_test,y_pred))
```
